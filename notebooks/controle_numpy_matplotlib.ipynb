# Partie F — Mini-Analyse Finale

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
# Votre code ici
#calcule de la matrice de corrélation

correlation_matrix = np.corrcoef(features, rowvar=False)
#affichage
print("Matrice de corrélation :\n", correlation_matrix)
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
# Votre code ici
#visualisation de la correlation la plus fort
plt.figure(figsize=(8, 6))
plt.scatter(features[:, 2],features[:, 3], c='orange', alpha=0.6)
#ajoute ligne tendance
m, b = np.polyfit(features[:, 2], features[:, 3], 1)
plt.plot(features[:, 2], m * features[:, 2] + b, color='red')
plt.title(" Visualisation de la Correlation (pétales) ")
plt.xlabel("Longueur des pétales(cm)")
plt.ylabel("Largeur des pétales(cm)")
plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# 1. Calcul de la variance pour chaque groupe de 50 lignes

var_setosa = np.mean(np.var(features[0:50], axis=0))
var_versicolor = np.mean(np.var(features[50:100], axis=0))
var_virginica = np.mean(np.var(features[100:150], axis=0))

variances = [var_setosa, var_versicolor, var_virginica]
especes = ['Setosa', 'Versicolor', 'Virginica']

# Identifier l'espèce avec la plus grande variance
index_max = np.argmax(variances)
print(f"L'espèce avec la plus grande variance globale est : {especes[index_max]}")

# 2. Visualisation des distributions
plt.figure(figsize=(10, 6))
plt.hist(features[0:50, 2], bins=10, alpha=0.5, label='Setosa', color='black')
plt.hist(features[50:100, 2], bins=10, alpha=0.5, label='Versicolor', color='blue')
plt.hist(features[100:150, 2], bins=10, alpha=0.5, label='Virginica', color='red')

plt.title("Distribution de la longueur des pétales par espèce")
plt.xlabel("Longueur (cm)")
plt.ylabel("Fréquence")
plt.legend()
plt.show()
```
